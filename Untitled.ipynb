```python
data = ['6,0\n', '37,10\n', '82,20\n', '127,30\n', '180,40\n', '258,50\n', '321,60\n', '347,70\n', '340,80\n', '342,90\n']

# カンマを基準に分割して右側を1/10にする
converted = []
for item in data:
    left, right = item.strip().split(',')
    new_right = int(int(right)/ 10)
    converted.append(f'{left},{new_right}')

print(converted)
```
